<h3 style = "color:green"> 3. GROUP BY and ORDER BY Clause


<div style = "background-color:lightblue; color:green; font-size:20px; text-align:justify">GROUP BY is used to group rows that have same values in one or more columns and then perform aggregate calculations.<br>
Aggregate Calculations: COUNT(), SUM(), AVG(), MIN(), MAX()

```python
%%sql

SELECT sex
FROM employees
GROUP BY sex;

```

```python
%%sql

SELECT sex, AVG(age), COUNT(age), MAX(age), MIN(age)
FROM employees
GROUP BY sex;
```

<h3 style = "color:green">3.3 HAVING Clause

<div style = "background-color:lightblue; color:green; font-size:20px; text-align:justify">
<ul>
<li>WHERE: It filters row before grouping and cannot use aggregate functions.</li>
<li>HAVING: It filters row after grouping and can use aggregate functioons.</li>
</ul>

</div>

```python
# %%sql

# SELECT sex, AVG(age)
# FROM employees
# WHERE AVG(age)>30
# GROUP BY sex;

#Gives error
```

```python
%%sql

SELECT sex, AVG(age)
FROM employees
GROUP BY sex
HAVING AVG(age)>30;
```

```python


%%sql

--With Aliasing

SELECT sex, AVG(age) AS avg_age
FROM employees
GROUP BY sex
HAVING avg_age > 30;
```

```python
%%sql

SELECT department, AVG(salary)
FROM employees
GROUP BY department
HAVING AVG(salary) >50000;
```

<h3 style = "color:green">5. JOIN Command

<div style = "background-color:lightblue; color:green; font-size:20px; text-align:justify"> A JOIN is used to combine rows from two or more tables based on a related column between them.

<h3 style = "color:green">5.1 INNER JOIN

<div style = "background-color:lightblue; color:green; font-size:20px; text-align:justify">INNER JOIN: Returns only matching rows.

```python
%%sql

SELECT *
FROM employees
JOIN employees_info
ON employees.id = employees_info.id
;
```

```python
%%sql

SELECT employees.id, name, employees.department
FROM employees
JOIN employees_info
ON employees.id = employees_info.id
;
```

<h3 style = "color:green">5.2 LEFT OUTER JOIN

<div style = "background-color:lightblue; color:green; font-size:20px; text-align:justify">LEFT OUTER JOIN: Returns all rows from the left table and matching rows from the right table.

```python
%%sql

SELECT *
FROM employees
LEFT JOIN employees_info
ON employees.dept_id = employees_info.dept_id
;
```

<h3 style = "color:green">5.3 RIGHT OUTER JOIN

<div style = "background-color:lightblue; color:green; font-size:20px; text-align:justify">RIGHT OUTER JOIN: Returns all the rows from the right table and matching rows from the left table.

```python
%%sql

SELECT *
FROM employees
RIGHT JOIN employees_info
ON employees.dept_id = employees_info.dept_id
;
```

<h3 style = "color:green">5.4 SELF JOIN

<div style = "background-color:lightblue; color:green; font-size:20px; text-align:justify">SELF JOIN: A self join is when a table is joined with itself. It is useful when rows in the same table are related to other rows in that same table.

```python
%%sql

SELECT *
FROM employees AS emp1
JOIN employees AS emp2
ON emp1.id = emp1.id
;
```

```python
%%sql


SELECT
    emp1.name AS Employee,
    emp2.name AS Manager
FROM employees AS emp1
JOIN employees AS emp2
ON emp1.id+1 = emp2.id
;
```
